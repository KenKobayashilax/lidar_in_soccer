```python
#make pcd file by using ipynb
import os
import get_frames.csv2pcd as csv2pcd
from settings import DATA_PATH

res = 100 #time resolution of frames(mili second)
start = 13.5 #the time(second) to set as frame0 <- this should be the time when human started to run.

data_type_name_csv = "LIVOX_Hallway_csv" #directry name of csv file
data_type_name_pcd = "LIVOX_Hallway_pcds" #directry name of pcd files

file_name="jogging_fast_4th" #name of file you use

# Replace 'your_input.csv' and 'output_directory' with your actual input CSV file and output directory
csv_file_path = os.path.join(DATA_PATH, data_type_name_csv, file_name+".csv")
# make save path corresponding to resolution and start timing.
output_dir = csv2pcd.make_save_dir_path(data_path=DATA_PATH, 
                       data_type_name=data_type_name_pcd, 
                       file_name=file_name, 
                       res=res, start=start)

print("making points list")
# Open the CSV file and read it line by line
points_by_timestamp = csv2pcd.read_csv(csv_file_path, res, start)
print("points list is made")  

#save each frames to pcd file
csv2pcd.save_points(points_by_timestamp, output_dir, res)
print("Done!")
```
