```python
#make pcd file by using ipynb
import configparser
import os
import get_frames.csv2pcd as csv2pcd

def read_config():
    config = configparser.ConfigParser()
    config.read("config.ini")
    return config

config = read_config()
# Access the DATA_PATH from the configuration
data_path = config.get("Paths", "DATA_PATH")

csv_file_path = fr"{data_path}/LIVOX_Hallway_csv/jogging_fast_4th.csv"
output_dir = fr"{data_path}/LIVOX_Hallway_pcds/jogging_fast_4th/res10ms"

res = 10 #time resolution of frames(mili second)

print("making points list")
# Open the CSV file and read it line by line
points_by_timestamp = csv2pcd.read_csv(csv_file_path, res)
print("points list is made")  

#save each frames to pcd file
csv2pcd.save_points(points_by_timestamp, output_dir, res)
print("Done!")
```
